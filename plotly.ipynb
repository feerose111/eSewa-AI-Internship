```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('/content/digital_wallet_transactions.csv')
```

```python
df.head()
```

```python
# Create an interactive scatter plot
fig = px.scatter(df, x="product_category", y="product_amount")
fig.show()
```

```python
# Ensure datetime format
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# Create new column for Month or Week
#df['Month'] = df['transaction_date'].dt.to_period('M').astype(str)  # For Month
df['Week'] = df['transaction_date'].dt.to_period('W').astype(str)  # For Week

# Plot by month
fig = px.line(df, x="transaction_date", y="product_amount", color="Week")
fig.show()
```

```python
#Total Sales by Product Category (Pie Chart)
fig = px.pie(
    df,
    names="product_category",
    values="product_amount",
    title="Sales Distribution by Category",
    hole=0.01
)
fig.show()
```

 Daily Transaction Trends (Line Plot)

```python
daily_sales = df.groupby("transaction_date")['product_amount'].sum().reset_index()
```

```python
print(daily_sales)
```

```python
fig = px.line(
    daily_sales,
    x="transaction_date",
    y="product_amount",
    title="Daily Sales Trend"
)
fig.update_xaxes(rangeslider_visible=True)  # Adds a date range slider
fig.show()
```

```python
#Cashback vs. Loyalty Points (Scatter Plot)
fig = px.scatter(
    df,
    x="cashback",
    y="loyalty_points",
    color="payment_method",
    size="product_amount",
    hover_name="user_id",
    title="Cashback vs. Loyalty Points"
)
fig.show()
```

```python
#Payment Method Analysis (Bar Chart)
fig = px.bar(
    df.groupby("payment_method")["transaction_id"].count().reset_index(),
    x="payment_method",
    y="transaction_id",
    color="payment_method",
    title="Transactions by Payment Method"
)
fig.show()
```

```python
#Merchant Performance (Sunburst Chart)
fig = px.sunburst(
    df,
    path=["merchant_name", "product_category"],
    values="product_amount",
    title="Merchant Sales Breakdown"
)
fig.show()
```
